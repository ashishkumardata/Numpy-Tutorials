```python
import numpy as np
```

```python
ages = np.array([[21,17,19,20,16,30,18,65],
                 [39,22,15,99,18,19,20,21]])
```

```python
teenagers = ages[ages < 18]
print(teenagers)
```

```python
adults = ages[(ages >= 18) & (ages < 65)]
print(adults)
```

```python
seniors = ages[ages >= 65]
print(seniors)
```

```python
evens = ages[ages % 2 == 0]
print(evens)
```

```python
odds = ages[ages % 2 != 0]
print(odds)
```

```python
adults = np.where(ages >= 18, ages, -1)
print(adults)
```

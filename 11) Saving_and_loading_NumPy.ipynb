```python
import numpy as np
```

```python
array = np.array([[1,2,3],[4,5,6]])
```

```python
np.save("data", array)
print("Numpy file was saved")
```

```python
#load file
arr = np.load("data.npy")
print("file loaded successfully")
```

```python
print(arr)
```

```python
# save multiply arrays
array1 = np.array([[1,2,3],[4,5,6]])
array2 = np.array([1.6,2.3,3.1,4.1])
```

```python
np.savez("datas", array1, array2)
print("saved")
```

```python
np.savez_compressed("compressed", array1, array2)
print("saved")
```

```python
# load multiple arrays
arrays = np.load("datas.npz")
```

```python
print(arrays)
```

```python
array1 = arrays["arr_0"]
array2 = arrays["arr_1"]
```

```python
print(array1)
print(array2)
```

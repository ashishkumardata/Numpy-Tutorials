```python
import numpy as np
```

```python
array = np.zeros(10)
print(array)
```

```python
array = np.zeros((2,3))
print(array)
```

```python
array = np.zeros((2,3,10))
print(array)
```

```python
array = np.ones((2,3,10))
print(array)
```

```python
array = np.full((2,3,10),5)
print(array)
```

```python
array = np.eye(3)
print(array)
```

```python
array = np.empty((2,3))
print(array)
```

```python
array = np.arange(1,10,2) # start stop step
print(array)
```

```python
array = np.arange(1,10,2.5)
print(array)
```

```python
array = np.linspace(1,10,4) # start stop num
print(array)
```

```python
array = np.linspace(1,10,6) # start stop num
print(array)
```

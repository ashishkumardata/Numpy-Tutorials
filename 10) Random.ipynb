```python
import numpy as np
```

```python
rng = np.random.default_rng()
```

```python
print(rng.integers(low= 1, high= 7))
```

```python
print(rng.integers(low= 1, high= 101))
```

```python
print(rng.integers(low= 1, high= 7, size= 3))
```

```python
print(rng.integers(low= 1, high= 7, size= (3,2)))
```

```python
rng = np.random.default_rng(seed = 1)
print(rng.integers(low= 1, high= 7, size= (3,2)))
```

```python
# Floating point numbers
print(np.random.uniform(low= -1, high= 1))
```

```python
print(np.random.uniform(low= -1, high= 1,size= 3))
```

```python
print(np.random.uniform(low= -1, high= 1,size= (3,2)))
```

```python
# shuffle
rng = np.random.default_rng()
array = np.array([1,2,3,4,5])
rng.shuffle(array)
print(array)
```

```python
# random choice
rng = np.random.default_rng()
fruits = np.array(["🍇","🍉","🍎","🥭","🍍"])
fruit = rng.choice(fruits)
print(fruit)
```

```python
fruits = rng.choice(fruits,size = 3)
print(fruits)
```

```python
fruits = rng.choice(fruits,size = (3,3))
print(fruits)
```

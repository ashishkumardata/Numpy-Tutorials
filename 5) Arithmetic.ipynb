```python
import numpy as np
```

```python
# scalar arithmetic
array = np.array([1,2,3])
```

```python
print(array + 1)
print(array - 2)
print(array * 3)
print(array / 4)
print(array ** 5)
```

```python
# vectorized math fuction
print(np.sqrt(array))
```

```python
arr = np.array([1.1,2.4,2.6])
print(np.round(arr))
```

```python
print(np.floor(arr))
```

```python
print(np.ceil(arr))
```

```python
print(np.pi)
```

```python
#Exercise
radii = np.array([1,2,3])
```

```python
# area of a circle
print(np.pi * radii ** 2)
```

```python
# element wise arithmetic
array1 = np.array([1,2,3])
array2 = np.array([4,5,6])
```

```python
print(array1 +array2)
print(array1 - array2)
print(array1 * array2)
print(array1 / array2)
print(array1 ** array2)
```

```python
# comparison operation
score = np.array([91,55,100,73,82,64])
```

```python
print(score == 100)
```

```python
print(score >= 60)
```

```python
print(score <= 60)
```

```python
score[score < 60] = 0
print(score)
```

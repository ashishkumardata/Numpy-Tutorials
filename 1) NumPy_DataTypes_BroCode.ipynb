```python
import numpy as np
```

```python
print(np.__version__)
```

```python
list1 = [1,2,3,4]
list1 = list1 * 2
print(list1)
print(type(list1))
```

```python
n1 = np.array([1,2,3,4])
n1 = n1 * 2 # also works n1 *=2
print(n1)
print(type(n1))
```

```python
n1 = np.array([1,2,3,4])
print(n1)
print(n1.dtype)
print(f"{n1.nbytes} bytes")
```

```python
n1 = np.array([1,2,3,4],dtype = np.int32)
print(n1)
print(n1.dtype)
print(f"{n1.nbytes} bytes")
```

```python
n1 = np.array([1,2,3,4],dtype = np.float64)
print(n1)
print(n1.dtype)
print(f"{n1.nbytes} bytes")
```

```python
n1 = np.array([1,2,3,4],dtype = np.bool_)
print(n1)
print(n1.dtype)
print(f"{n1.nbytes} bytes")
```

```python
n1 = np.array([1,2,3,4],dtype = np.str_)
print(n1)
print(n1.dtype)
print(f"{n1.nbytes} bytes")
```

```python
n1 = np.array([1000,2000,3000,4000],dtype = np.str_)
print(n1)
print(n1.dtype)
print(f"{n1.nbytes} bytes")
```

```python
n1 = np.array([1000,2000,3000,4000],dtype = "<U3")
print(n1)
print(n1.dtype)
print(f"{n1.nbytes} bytes")
```

```python
n1 = np.array([1,"string",2,True,3.5,4],dtype = np.object_)
print(n1)
print(n1.dtype)
print(f"{n1.nbytes} bytes")
```

```python
n1 = np.array([1.,2., 3., 4.])

n1 = n1.astype(np.int16)

print(n1)
print(n1.dtype)
print(f"{n1.nbytes} bytes")
```
